```python
import numpy as np
```

```python
# ejercicio 1 Dado el arreglo a = np.array([1, 2, 3, 4]), determina si ninguno de los elementos es cero.
a = np.array([1, 2, 3, 4])
resultado = np.all(a != 0)  
print(resultado)  
```

```python
# ejercicio 2 Dado el arreglo a = np.array([0, 0, 2, 0]), determina si contiene algún elemento que no sea cero.
a = np.array([0, 0, 2, 0])
resultado = np.any(a != 0)
print(resultado)  
```

```python
# ejercicio 3 
# Analiza el arreglo a = np.array([1, 2, np.nan, np.inf]) y 
# determina si contiene valores que sean infinitos o valores no numéricos (NaN).
a = np.array([1, 2, np.nan, np.inf])

# 1. Verificar si contiene infinitos (positivos o negativos)
tiene_inf = np.any(np.isinf(a))
print("¿Contiene infinitos?:", tiene_inf)  

# 2. Verificar si contiene NaN
tiene_nan = np.any(np.isnan(a))
print("¿Contiene NaN?:", tiene_nan)  

```

```python
# ejercicio 4 Dado el arreglo a = np.array([1, 2, np.inf]), determina si contiene valores infinitos.
a = np.array([1, 2, np.inf])
tiene_inf = np.any(np.isinf(a))
print("¿Contiene infinitos?:", tiene_inf)  
```

```python
# ejercicio 5 Dado el arreglo a = np.array([1, np.nan, 3]), determina si contiene valores NaN.

a = np.array([1, np.nan, 3])
tiene_nan = np.any(np.isnan(a))
print("¿Contiene NaN?:", tiene_nan)
```

```python
# ejercicio 6 Dados los arreglos a = np.array([1, 3, 5]) y b = np.array([2, 3, 4]), realiza una comparación elemento a elemento 
# contemplando mayor, mayor o igual, menor y menor o igual.

a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

print("Mayor:", a > b)
print("Mayor o igual:", a >= b)
print("Menor:", a < b)
print("Menor o igual:", a <= b)
```

```python
# ejercicio 7 Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.
arreglo = np.array([1, 7, 13, 105])
print("Memoria ocupada (bytes):", arreglo.nbytes)
```

```python
# ejercicio 8 Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.
ceros = np.zeros(10)
unos = np.ones(10)
cincos = np.full(10, 5)
print("Arreglo de ceros:", ceros)
print("Arreglo de unos:", unos)
print("Arreglo de cincos:", cincos)
```

```python
# ejercicio 9 Crea un arreglo que contenga todos los números enteros entre 30 y 70.
arreglo = np.arange(30, 71)
print(arreglo)
```

```python
# ejercicio 10 Crea un arreglo que contenga los números pares entre 30 y 70.
arreglo_pares = np.arange(30, 71, 2)
print(arreglo_pares)
```

```python
# ejercicio 11 Crea una matriz identidad de 3x3.
matriz_identidad = np.identity(3)
print(matriz_identidad)
```

```python
# ejercicio 12 Genera un número aleatorio entre 0 y 1.
numero_aleatorio = np.random.rand()
print(numero_aleatorio)
```

```python
# ejercicio 13 Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
arreglo_aleatorio = np.random.randn(15)
print(arreglo_aleatorio)
```

```python
# ejercicio 12 Crea un arreglo con valores entre 15 y 55, e imprime todos los valores excepto el primero y el último.
arreglo = np.arange(15, 56)
print(arreglo[1:-1])  # Imprime todos los elementos excepto el primero y el último
```

```python
# ejercicio 13 Crea un vector (arreglo) de longitud 10 con valores equitativamente distribuidos entre 5 y 50.
vector = np.linspace(5, 50, 10)
print(vector)
```

```python
# ejercicio 16 Crea un vector con valores de 0 a 20, y cambia el signo de los números de 9 a 15.
vector = np.arange(21)
vector = np.where((vector >= 9) & (vector <= 15), -vector, vector)
print(vector)
```

```python
# ejercicio 17 Convierte el arreglo a = np.array([1, 2, 3]) a una lista de Python.
a = np.array([1, 2, 3])
b = a.tolist()

print("Tipo:", type(a))
print("Tipo:", type(b))
```

```python
# ejercicio 18 Convierte la lista l = [1, 2, 3] a un arreglo de Numpy.
l = [1, 2, 3]
a = np.array(l)
print("Tipo:", type(l))
print("Tipo:", type(a))
```

```python
# ejercicio 19 Dada la matriz m = np.array([[1, 2], [3, 4]]), verifica qué números son pares y almacena los resultados en otra matriz.
a = np.array([[1, 2], [3, 4]])
pares = (a % 2 == 0)
print(pares)

# alternativa
resultado_str = np.where(a % 2 == 0, "Par", "Impar")
print("\nMatriz con strings:")
print(resultado_str)

```

```python
# ejercicio 20 Dado el arreglo m = np.array([[1, 2, 3], [4, 5, 6]]), encuentra el número de filas y columnas de la matriz.
m = np.array([[1, 2, 3], [4, 5, 6]])
filas, columnas = m.shape
print("Número de filas:", filas)
print("Número de columnas:", columnas)
```
